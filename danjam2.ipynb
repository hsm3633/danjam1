```python
import random

# --- ê°ì • ë°ì´í„° ì…ë ¥ ë° ê³„ì‚° --- #
def get_user_emotion_data():
    """
    Collects user input data in Korean and calculates a weighted score based on their inputs.
    Returns the overall weighted emotion score.
    """
    print("\n--- ê°ì • ë°ì´í„° ì…ë ¥ ---")
    color = input("ì˜¤ëŠ˜ì˜ ê¸°ë¶„ì„ ë‚˜íƒ€ë‚´ëŠ” ìƒ‰ìƒì„ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: íŒŒë‘, ë…¸ë‘, ë¹¨ê°•, ê²€ì •, ì´ˆë¡, íšŒìƒ‰): ").strip()
    music = input("ì˜¤ëŠ˜ ë“¤ì€ ìŒì•…ì˜ ë¶„ìœ„ê¸°ë¥¼ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: í–‰ë³µ, ìŠ¬í””, ì°¨ë¶„, ì‹ ë‚¨, ìš°ìš¸): ").strip()
    emoji = input("í˜„ì¬ ê¸°ë¶„ì„ ë‚˜íƒ€ë‚´ëŠ” ì´ëª¨ì§€ë¥¼ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: ğŸ˜Š,ğŸ˜Œ,ğŸ˜¡,ğŸ˜”,ğŸ˜¢): ").strip()
   
    # ì ìˆ˜ í…Œì´ë¸”
    color_scores = {"íŒŒë‘": 2, "ë…¸ë‘": 5, "ë¹¨ê°•": 5, "ì´ˆë¡": 4, "ê²€ì •": 1, "íšŒìƒ‰": 1}
    music_scores = {"í–‰ë³µ": 5, "ìŠ¬í””": 2, "ì°¨ë¶„": 4, "ì‹ ë‚¨": 5, "ìš°ìš¸": 1}
    emoji_scores = {"ğŸ˜Š": 5, "ğŸ˜Œ": 4, "ğŸ˜¡": 3, "ğŸ˜”": 2, "ğŸ˜¢": 1}
   
    # ì ìˆ˜ ê³„ì‚°
    score_color = color_scores.get(color, 3)
    score_music = music_scores.get(music, 3)
    score_emoji = emoji_scores.get(emoji, 3)
   
    # ê°€ì¤‘ì¹˜ ë¶€ì—¬ ë° ì ìˆ˜ ê³„ì‚°
    weight_color = 1
    weight_music = 1
    weight_emoji = 4
   
    weighted_score = (
        score_color * weight_color +
        score_music * weight_music +
        score_emoji * weight_emoji
    )
    normalized_score = round(weighted_score / 10, 1)
    return normalized_score

# --- ê°ì • ìƒíƒœ ë¶„ë¥˜ --- #
def classify_emotion_state(score):
    if score <= 1.5:
        return "ë§¤ìš° ì¢‹ì§€ ì•ŠìŒ"
    elif score <= 2.5:
        return "ì¢‹ì§€ ì•ŠìŒ"
    elif score <= 3.5:
        return "ë³´í†µ"
    elif score <= 4.5:
        return "ì¢‹ìŒ"
    else:
        return "ë§¤ìš° ì¢‹ìŒ"

# --- PHQ-9 ì§ˆë¬¸ --- #
PHQ9_FIXED_QUESTIONS = [
    "ì¼ìƒì ì¸ ì¼ì— ëŒ€í•œ í¥ë¯¸ë‚˜ ì¦ê±°ì›€ì´ ê±°ì˜ ì—†ë‹¤.",
    "ì ë“¤ê¸° ì–´ë µê±°ë‚˜ ìì£¼ ê¹¨ê±°ë‚˜ ë„ˆë¬´ ë§ì´ ì”ë‹¤.",
    "ê¸°ë¶„ì´ ê°€ë¼ì•‰ê±°ë‚˜ ìš°ìš¸í•˜ê±°ë‚˜ í¬ë§ì´ ì—†ë‹¤ëŠ” ëŠë‚Œì´ ë“ ë‹¤."
]

PHQ9_RANDOM_QUESTIONS = [
    "ì‹ìš•ì´ ì¤„ì—ˆê±°ë‚˜ ì§€ë‚˜ì¹˜ê²Œ ë§ì´ ë¨¹ëŠ”ë‹¤.",
    "í”¼ê³¤í•˜ê±°ë‚˜ ê¸°ìš´ì´ ê±°ì˜ ì—†ë‹¤.",
    "ìì‹ ì— ëŒ€í•œ ì¢‹ì§€ ì•Šì€ í‰ê°€(ë‚´ê°€ ì‹¤íŒ¨ìë¼ê³  ëŠë¼ê±°ë‚˜ ë‚˜ ìì‹ ì´ë‚˜ ê°€ì¡±ì„ ì‹¤ë§ì‹œì¼°ë‹¤ê³  ëŠë‚€ë‹¤).",
    "ì§‘ì¤‘í•˜ëŠ” ë° ì–´ë ¤ì›€ì„ ê²ªëŠ”ë‹¤(ì˜ˆ: ì‹ ë¬¸ ì½ê¸°ë‚˜ TV ë³´ê¸°).",
    "ë‹¤ë¥¸ ì‚¬ëŒë“¤ì´ ì•Œì•„ì°¨ë¦´ ì •ë„ë¡œ ë„ˆë¬´ ëŠë ¤ì¡Œê±°ë‚˜ ì´ˆì¡°í•˜ê±°ë‚˜ ì•ˆì ˆë¶€ì ˆëª»í•˜ëŠ” ëŠë‚Œì´ ë“ ë‹¤.",
    "ì°¨ë¼ë¦¬ ì£½ëŠ” ê²ƒì´ ë‚«ê² ë‹¤ê³  ìƒê°í•˜ê±°ë‚˜ ì–´ë–¤ ì‹ìœ¼ë¡œë“  ìì‹ ì„ í•´ì¹˜ê³  ì‹¶ë‹¤ëŠ” ìƒê°ì„ í•œë‹¤."
]

def run_phq9_questionnaire(fixed_questions, random_questions):
    """
    Asks fixed and random PHQ-9 questions, collects O/X responses, and calculates the score.
    """
    total_score = 0
    questions = fixed_questions + random.sample(random_questions, 2)

    for question in questions:
        while True:
            response = input(f"{question} (O/X): ").strip().upper()
            if response == "O":
                total_score += 1
                break
            elif response == "X":
                break
            else:
                print("ì˜ëª»ëœ ì…ë ¥ì…ë‹ˆë‹¤. O ë˜ëŠ” Xë¡œë§Œ ì‘ë‹µí•´ì£¼ì„¸ìš”.")
    return total_score

def run_phq9_questionnaire2(fixed_questions, random_questions):
    """
    Asks fixed and random PHQ-9 questions, collects responses, and calculates the score.
    """
    total_score = 0
    questions = fixed_questions + random.sample(random_questions, 6)

    for question in questions:
        while True:
            response = input(f"{question} (0: ì „í˜€ ì—†ìŒ, 1: ë©°ì¹  ë™ì•ˆ, 2: ì¼ì£¼ì¼ ì´ìƒ, 3: ê±°ì˜ ë§¤ì¼): ").strip()
            if response in ["0", "1", "2", "3"]:
                total_score += int(response)
                break
            else:
                print("ì˜ëª»ëœ ì…ë ¥ì…ë‹ˆë‹¤. 0, 1, 2, 3 ì¤‘ í•˜ë‚˜ë¡œ ì‘ë‹µí•´ì£¼ì„¸ìš”.")
    return total_score

# --- 2ì°¨ ê²€ì¦ ì„¤ë¬¸ --- #
def second_validation_survey():
    """
    Conducts a second validation survey for moderate to severe depression symptoms.
    """
    print("\n--- 2ì°¨ ê²€ì¦ ì„¤ë¬¸ ---")
    return run_phq9_questionnaire2(PHQ9_FIXED_QUESTIONS, PHQ9_RANDOM_QUESTIONS)

# --- í”¼ë“œë°± ì œê³µ --- #
def provide_feedback(score):
    """
    Provides feedback based on the PHQ-9 score.
    """
    if score <= 4:
        return "ìš°ìš¸ ì¦ìƒ ì—†ìŒ. ê¸ì •ì ì¸ ìƒíƒœë¥¼ ìœ ì§€í•˜ì„¸ìš”!"
    elif score <= 9:
        return "ê²½ë¯¸í•œ ìš°ìš¸ ì¦ìƒ. ê°„ë‹¨í•œ íƒœìŠ¤í¬ë¥¼ ìˆ˜í–‰í•´ë³´ì„¸ìš”."
    elif score <= 14:
        return "ì¤‘ê°„ ì •ë„ì˜ ìš°ìš¸ ì¦ìƒ. ì¶”ê°€ ì„¤ë¬¸ì„ ì§„í–‰í•˜ì„¸ìš”."
    elif score <= 19:
        return "ì‹¬í•œ ìš°ìš¸ ì¦ìƒ. ì „ë¬¸ê°€ì˜ ìƒë‹´ì´ ê¶Œì¥ë©ë‹ˆë‹¤."
    else:
        return "ë§¤ìš° ì‹¬í•œ ìš°ìš¸ ì¦ìƒ. ì¦‰ê°ì ì¸ ë„ì›€ì„ ìš”ì²­í•˜ì„¸ìš”."

# --- ë©”ì¸ ì‹¤í–‰ í•¨ìˆ˜ --- #
def main():
    total_phq9_score = 0

    for i in range(14):
        print(f"\n--- {i + 1}/14ë²ˆì§¸ ì„¤ë¬¸ ---")
        
        # ê°ì • ë°ì´í„° ì…ë ¥
        emotion_score = get_user_emotion_data()
        print(f"Emotion Score: {emotion_score}")

        # ê°ì • ìƒíƒœ ë¶„ë¥˜
        state = classify_emotion_state(emotion_score)
        print(f"Emotion State: {state}")

        # PHQ-9 ì„¤ë¬¸ ì¡°ê±´
        if state in ["ì¢‹ì§€ ì•ŠìŒ", "ë§¤ìš° ì¢‹ì§€ ì•ŠìŒ"]:
            phq9_score = run_phq9_questionnaire(PHQ9_FIXED_QUESTIONS, PHQ9_RANDOM_QUESTIONS)
            total_phq9_score += phq9_score

    # ê²°ê³¼ ì¶œë ¥
    if total_phq9_score >= 42:
        score = second_validation_survey()
        feedback = provide_feedback(score)
        print(f"2ì°¨ ê²€ì¦ ì ìˆ˜: {score}")
        print(f"Feedback: {feedback}")

# ì‹¤í–‰
main()
```
